#### 시군구별 인구 데이터

```python
po = pd.read_csv('./population.csv')
po.head()
```

```python
# 필요없는 columns 삭제
po = po[["소재지(시군구)별(1)","소재지(시군구)별(2)","2020"]]
po.head()
```

```python
# 필요없는 row 삭제
po = po.drop([0,1], 'index')
po.head()
```

```python
# 소재시(시군구)별(2)에서 값이 '소계'인 행 제거
poo = po['소재지(시군구)별(2)'].isin(['소계'])
po = po[~poo]
po.head()
```

```python
# reset index
po = po.reset_index()
po = po.drop('index', 'columns')
po.head()
```

```python
### 데이터 엑셀로 내보내기
po.to_excel('population_by_sigoongu.xlsx')
```

### 두 데이터 join하기
#### 한 테이블에서는 세종시가 세종특별자치시로 입력되어있음
##### si_goon_gu에서는 세종특별자치시 세종시, po~에서는 세종특별자치시 세종특별자치시

#### [ 엑셀에서 데이터 합치기 ]
##### 1. si_goon_gu 데이터에서 ="시도명칭"&" "&"시군구명칭2" 합쳐서 새 컬럼 만들기
##### 2. po 데이터에서(population_by_sigoongu에서) ="소재지(시군구)별(1)"&" "&"소재지(시군구)별(2)" 합쳐서 새 컬럼 만들기
##### 3. =XLOOKUP(si_goon_gu 새 컬럼 데이터 하나, po~ 데이터 새 컬럼, po~데이터 2020(인구 수) 컬럼

```python
# 두 데이터 간에 세종시 호칭 세종특별자치시로 통일하기
si_goon_gu = si_goon_gu.replace('세종시','세종특별자치시')
```

```python
# si_goon_gu 데이터에 "시도명칭"&" "&"시군구명칭2" 합쳐서 새 컬럼 만들기
cols = ['시도명칭', '시군구명칭2']
si_goon_gu['시군구명_합치기'] = si_goon_gu[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
si_goon_gu
```

```python
po = po.rename(columns={"2020":"총 인구수"})

# po 데이터에서 "소재지(시군구)별(1)"&" "&"소재지(시군구)별(2)" 합쳐서 새 컬럼 만들기
col = ['소재지(시군구)별(1)', '소재지(시군구)별(2)']
po['소재지_합치기'] = po[col].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
po
```

```python
# si_goon_gu '시군구명_합치기', po 데이터 '소재지_합치기', '총 인구수' 컬럼 join하기
sisi = pd.merge(si_goon_gu, po, 'left', left_on='시군구명_합치기', right_on='소재지_합치기')
sisi
```

```python
### 데이터 다듬기
population_sigoongu = sisi[["시도코드","시도명칭","시군구코드","시군구명칭","시군구명칭2","총 인구수"]]
population_sigoongu.head()
```

```python
### 데이터 엑셀로 내보내기
population_sigoongu.to_excel('population_sigoongu.xlsx')
```
